```python
import matplotlib.pyplot as plt
import networkx as nx
import random
```

# generatie adjacency_matirix

```python
graph = nx.generators.random_graphs.gnm_random_graph(n=100, m=200)
```

```python
adjacency_matrix = nx.adjacency_matrix(graph).todense()
adjacency_matrix
```

# check undirected

```python
def symmetrical(matrix):
    simetrica=True
    for i in range(0, len(matrix)):
        for j in range(0,len(matrix)):
            if (matrix[i][j] != matrix[j][i]):
                simetrica=False
    return simetrica
```

```python
symmetrical(adjacency_matrix)
```

Use Depth-first search to find connected components of the graph

```python
connected_components = []

for component in nx.connected_components(graph):
    connected_components.append(component)

for i, component in enumerate(connected_components):
    print(f"Connected Component {i + 1}: {component}")
```

Shortest routine

```python
start_node = random.randint (0,100)
end_node = random.randint (0,100)
shortest_path = nx.shortest_path(graph, source=start_node, target=end_node)
print(f"Shortest Path from Node {start_node} to Node {end_node}: {shortest_path}")
#visilisize
def painter_transform(path):
    edge_path =[]
    for  i in range(0,len(path[:-1])):
       edge_path.append((path[i],path[i+1]))
    return edge_path
pos = nx.circular_layout(graph)
plt.figure(figsize=(12,12))
nx.draw_networkx(graph, pos, font_size=8, node_size=200, node_color='tab:blue')
nx.draw_networkx_edges(graph, pos, edgelist=painter_transform(shortest_path), edge_color='yellow', width=2.0)
nx.draw_networkx_nodes(graph, pos, nodelist=shortest_path, node_size=200, node_color='red')
plt.show()
```
